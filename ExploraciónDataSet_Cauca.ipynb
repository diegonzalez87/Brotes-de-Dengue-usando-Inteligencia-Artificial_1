# 1. EXPLORACI√ìN DE DATOS EPIDEMIOL√ìGICOS

```python
# =============================================================================
# 1. EXPLORACI√ìN DE DATOS EPIDEMIOL√ìGICOS
# =============================================================================

print("="*80)
print("üìä AN√ÅLISIS DE DATOS EPIDEMIOL√ìGICOS (cases_clean.csv)")
print("="*80)

# Cargar datos de casos
cases_df = pd.read_csv('/content/cases_clean.csv')

print(f"\nüîç INFORMACI√ìN B√ÅSICA:")
print(f"   ‚Ä¢ Filas: {cases_df.shape[0]:,}")
print(f"   ‚Ä¢ Columnas: {cases_df.shape[1]}")
print(f"   ‚Ä¢ Per√≠odo: {cases_df.columns}")

print(f"\nüìã ESTRUCTURA DEL DATASET:")
print(f"   ‚Ä¢ Columnas disponibles:")
for i, col in enumerate(cases_df.columns, 1):
    print(f"     {i:2d}. {col}")

print(f"\nüìä INFORMACI√ìN DETALLADA POR COLUMNA:")
print(cases_df.info())

print(f"\nüéØ PRIMERAS 5 FILAS:")
print(cases_df.head())

print(f"\nüìà ESTAD√çSTICAS DESCRIPTIVAS:")
print(cases_df.describe())

# Buscar columnas de tiempo
time_columns = [col for col in cases_df.columns
                if any(keyword in col.lower() for keyword in ['date', 'tiempo', 'semana', 'mes', 'a√±o', 'year', 'week', 'month'])]
print(f"\nüìÖ COLUMNAS TEMPORALES DETECTADAS: {time_columns}")

# Buscar columnas geogr√°ficas
geo_columns = [col for col in cases_df.columns
               if any(keyword in col.lower() for keyword in ['municipio', 'municipality', 'lat', 'lon', 'coord', 'geo'])]
print(f"üìç COLUMNAS GEOGR√ÅFICAS DETECTADAS: {geo_columns}")

# Valores √∫nicos en columnas categ√≥ricas
categorical_cols = cases_df.select_dtypes(include=['object']).columns
print(f"\nüè∑Ô∏è  VALORES √öNICOS EN COLUMNAS CATEG√ìRICAS:")
for col in categorical_cols:
    unique_vals = cases_df[col].nunique()
    print(f"   ‚Ä¢ {col}: {unique_vals} valores √∫nicos")
    if unique_vals <= 20:  # Mostrar valores if son pocos
        print(f"     Valores: {list(cases_df[col].unique())[:10]}")
```

```python
# =============================================================================
# 2. EXPLORACI√ìN DE DATOS ENTOMOL√ìGICOS
# =============================================================================

print("\n" + "="*80)
print("ü¶ü AN√ÅLISIS DE DATOS ENTOMOL√ìGICOS (entomologico_clean.csv)")
print("="*80)

# Cargar datos entomol√≥gicos
#ento_df = pd.read_csv('/kaggle/input/stratification-of-dengue-in-cauca-colombia/entomologico_clean.csv')
ento_df = pd.read_csv('/content/entomologico_clean.csv')

print(f"\nüîç INFORMACI√ìN B√ÅSICA:")
print(f"   ‚Ä¢ Filas: {ento_df.shape[0]:,}")
print(f"   ‚Ä¢ Columnas: {ento_df.shape[1]}")

print(f"\nüìã COLUMNAS ENTOMOL√ìGICAS:")
for i, col in enumerate(ento_df.columns, 1):
    print(f"     {i:2d}. {col}")

print(f"\nüìä INFORMACI√ìN DETALLADA:")
print(ento_df.info())

print(f"\nüéØ PRIMERAS 5 FILAS:")
print(ento_df.head())

print(f"\nüìà ESTAD√çSTICAS DESCRIPTIVAS:")
print(ento_df.describe())

# Buscar √≠ndices entomol√≥gicos est√°ndar
entomo_indices = ['casa', 'breteau', 'container', 'indice', 'larvae', 'pupae', 'adult', 'aedes']
detected_indices = []
for col in ento_df.columns:
    for idx in entomo_indices:
        if idx.lower() in col.lower():
            detected_indices.append(col)
            break

print(f"\nü¶ü POSIBLES √çNDICES ENTOMOL√ìGICOS DETECTADOS:")
for idx in set(detected_indices):
    print(f"   ‚Ä¢ {idx}")

# Buscar columnas temporales en datos entomol√≥gicos
ento_time_cols = [col for col in ento_df.columns
                  if any(keyword in col.lower() for keyword in ['date', 'tiempo', 'semana', 'mes', 'a√±o', 'year', 'week', 'month'])]
print(f"\nüìÖ COLUMNAS TEMPORALES EN DATOS ENTOMOL√ìGICOS: {ento_time_cols}")
```

```python
# =============================================================================
# 3. EXPLORACI√ìN DE DATOS METEOROL√ìGICOS (NetCDF)
# =============================================================================

print("\n" + "="*80)
print("üå§Ô∏è  AN√ÅLISIS DE DATOS METEOROL√ìGICOS (H_mswx_2015_2022.nc)")
print("="*80)

try:
    # Cargar archivo NetCDF
    #weather_ds = xr.open_dataset('/kaggle/input/stratification-of-dengue-in-cauca-colombia/H_mswx_2015_2022.nc')
    #ento_df = pd.read_csv('/content/entomologico_clean.csv')
    weather_ds = xr.open_dataset('/content/H_mswx_2015_2022.nc')

    print(f"\nüîç INFORMACI√ìN GENERAL:")
    print(f"   ‚Ä¢ Dimensiones: {dict(weather_ds.dims)}")
    print(f"   ‚Ä¢ Variables: {list(weather_ds.data_vars)}")
    print(f"   ‚Ä¢ Coordenadas: {list(weather_ds.coords)}")

    print(f"\nüìä ESTRUCTURA COMPLETA:")
    print(weather_ds)

    print(f"\nüìÖ INFORMACI√ìN TEMPORAL:")
    if 'time' in weather_ds.coords:
        time_range = weather_ds.time
        print(f"   ‚Ä¢ Inicio: {str(time_range.values[0])[:10]}")
        print(f"   ‚Ä¢ Fin: {str(time_range.values[-1])[:10]}")
        print(f"   ‚Ä¢ Total per√≠odos: {len(time_range)}")
        print(f"   ‚Ä¢ Frecuencia estimada: {pd.infer_freq(pd.to_datetime(time_range.values[:100]))}")

    print(f"\nüåç INFORMACI√ìN ESPACIAL:")
    if 'lat' in weather_ds.coords and 'lon' in weather_ds.coords:
        lat_range = weather_ds.lat.values
        lon_range = weather_ds.lon.values
        print(f"   ‚Ä¢ Latitud: {lat_range.min():.3f} a {lat_range.max():.3f}")
        print(f"   ‚Ä¢ Longitud: {lon_range.min():.3f} a {lon_range.max():.3f}")
        print(f"   ‚Ä¢ Resoluci√≥n lat: ~{np.diff(lat_range).mean():.3f}¬∞")
        print(f"   ‚Ä¢ Resoluci√≥n lon: ~{np.diff(lon_range).mean():.3f}¬∞")
        print(f"   ‚Ä¢ Puntos de grilla: {len(lat_range)} x {len(lon_range)}")

    print(f"\nüå°Ô∏è VARIABLES METEOROL√ìGICAS:")
    for var_name, var_data in weather_ds.data_vars.items():
        print(f"   ‚Ä¢ {var_name}:")
        print(f"     - Dimensiones: {var_data.dims}")
        print(f"     - Forma: {var_data.shape}")
        if hasattr(var_data, 'attrs'):
            if 'units' in var_data.attrs:
                print(f"     - Unidades: {var_data.attrs['units']}")
            if 'long_name' in var_data.attrs:
                print(f"     - Descripci√≥n: {var_data.attrs['long_name']}")

        # Estad√≠sticas b√°sicas de la variable
        try:
            var_stats = var_data.values
            if var_stats.size > 0:
                print(f"     - Rango: {np.nanmin(var_stats):.2f} a {np.nanmax(var_stats):.2f}")
                print(f"     - Media: {np.nanmean(var_stats):.2f}")
        except:
            print("     - No se pudieron calcular estad√≠sticas")

    # Intentar identificar variables clim√°ticas clave
    climate_vars = {
        'temperatura': ['temp', 'temperature', 't2m', 'tas', 'tmax', 'tmin'],
        'precipitaci√≥n': ['precip', 'precipitation', 'prcp', 'pr', 'rain'],
        'humedad': ['humid', 'humidity', 'rh', 'hurs', 'rhum'],
        'viento': ['wind', 'viento', 'ws', 'was', 'u10', 'v10'],
        'presi√≥n': ['pressure', 'press', 'psl', 'ps', 'presion']
    }

    print(f"\nüéØ VARIABLES CLIM√ÅTICAS IDENTIFICADAS:")
    identified_vars = {}
    for climate_type, possible_names in climate_vars.items():
        found_vars = []
        for var_name in weather_ds.data_vars:
            if any(possible in var_name.lower() for possible in possible_names):
                found_vars.append(var_name)
        if found_vars:
            identified_vars[climate_type] = found_vars
            print(f"   ‚Ä¢ {climate_type.upper()}: {found_vars}")

except Exception as e:
    print(f"‚ùå Error al cargar datos meteorol√≥gicos: {e}")
    print("   Verifica que xarray est√© instalado: pip install xarray netCDF4")
```

```python
# =============================================================================
# 4. AN√ÅLISIS DE COMPATIBILIDAD TEMPORAL
# =============================================================================

print("\n" + "="*80)
print("‚è∞ AN√ÅLISIS DE COMPATIBILIDAD TEMPORAL ENTRE DATASETS")
print("="*80)

# Funci√≥n para extraer informaci√≥n temporal de cada dataset
def analyze_temporal_coverage(df, name, time_columns):
    """Analiza la cobertura temporal de un dataset"""
    print(f"\nüìÖ {name}:")

    if not time_columns:
        print("   ‚ùå No se detectaron columnas temporales claras")
        return None

    for time_col in time_columns:
        if time_col in df.columns:
            try:
                # Intentar convertir a datetime
                dates = pd.to_datetime(df[time_col], errors='coerce')
                valid_dates = dates.dropna()

                if len(valid_dates) > 0:
                    print(f"   ‚Ä¢ Columna: {time_col}")
                    print(f"   ‚Ä¢ Rango: {valid_dates.min()} a {valid_dates.max()}")
                    print(f"   ‚Ä¢ Registros v√°lidos: {len(valid_dates)} / {len(dates)}")

                    # Analizar frecuencia
                    if len(valid_dates) > 1:
                        freq = pd.infer_freq(valid_dates.sort_values()[:100])
                        print(f"   ‚Ä¢ Frecuencia inferida: {freq}")

                    return valid_dates.min(), valid_dates.max()
            except:
                print(f"   ‚ö†Ô∏è  No se pudo procesar {time_col} como fecha")

    return None

# Analizar cobertura temporal de cada dataset
cases_temporal = analyze_temporal_coverage(cases_df, "DATOS EPIDEMIOL√ìGICOS", time_columns)
ento_temporal = analyze_temporal_coverage(ento_df, "DATOS ENTOMOL√ìGICOS", ento_time_cols)

# Comparar con datos meteorol√≥gicos
if 'weather_ds' in locals():
    print(f"\nüìÖ DATOS METEOROL√ìGICOS:")
    if 'time' in weather_ds.coords:
        weather_start = pd.to_datetime(str(weather_ds.time.values[0]))
        weather_end = pd.to_datetime(str(weather_ds.time.values[-1]))
        print(f"   ‚Ä¢ Rango: {weather_start.date()} a {weather_end.date()}")
        weather_temporal = (weather_start, weather_end)
    else:
        weather_temporal = None
```
